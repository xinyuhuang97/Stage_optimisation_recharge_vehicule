```python
list_nb=[20, 40, 60, 80 ,150, 200]
list_gap=[]
list_gap_score=[]
for nb_ev in list_nb:
    nk=20
    n_pre=50
    K=50
    np.random.seed(0)
    instance_json(nb_ev)
    my_instance = "../data/instance_" + str(nb_ev) + ".json"
    standard = Frank_Wolfe_Standard_v2(my_instance)
    np.random.seed(0)
    x_bar_k,df = standard.FW_st_solve(actual_time, analyse=True, K=K, nk=nk,n_pre=n_pre,gap_calculate=True)
    list_gap.append(df['gap_primal_dual'].values)
    list_gap_score.append(df['best_score'].values)
```

```python
list_nb=[20, 40, 60, 80 ,150, 200]
for nb_ev in list_nb:
    my_instance = "../data/instance_" + str(nb_ev) + ".json"
    frontal = Frontal(my_instance,actual_time)
    np.random.seed(0)
    result,s_t_min,soc_max = frontal.Frontal_solve(actual_time)
```

```python
fig, axes = plt.subplots(len(list_nb)//3,3, figsize=(16, 8 ))

cpt=0
# Plot on each subplot
axes_y=0
for i,nb in enumerate(list_nb):
    np.random.seed(0)
    my_instance = "../data/instance_" + str(nb) + ".json"
    frontal = Frontal(my_instance,actual_time)
    result,s_t_min,soc_max = frontal.Frontal_solve(actual_time)
    axes[cpt][axes_y].set_title("Evolution of the objective function value with "+str(nb)+" vehicles")
    axes[cpt][axes_y].set_xlabel("Iteration")
    axes[cpt][axes_y].set_ylabel("Objective function value")
    axes[cpt][axes_y].plot([x for x in range(50)],list_gap_score[i],label=str(nb)+"vehicles")
    axes[cpt][axes_y].plot([x for x in range(50)],[frontal.objective_function(result) for x in range(50)],label="Cplex")
    axes[cpt][axes_y].legend()
    if (cpt-1)%3==0:
        axes_y+=1
        cpt=0
    else:
        cpt+=1

# Adjust spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```
