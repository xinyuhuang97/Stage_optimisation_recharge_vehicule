```python
N=200
actual_time=0
nk=20
n_pre=50
K=200
list_v2=np.zeros(K)
list_v3=np.zeros(K)

nb_test=10
for i in range(nb_test):
    np.random.seed(i)
    instance_json(N)
    my_instance = "../data/instance_" + str(N) + ".json"
    np.random.seed(0)
    frontal = Frontal(my_instance,actual_time)
    _, _, _, optimal_value = frontal.Frontal_solve(actual_time)
    standard_v2 = Frank_Wolfe_Standard_v2(my_instance)
    np.random.seed(0)
    x_bar_k_v2,df_v2=standard_v2.FW_st_solve(actual_time, analyse=True, K=K, nk=nk,n_pre=n_pre)
    list_v2+=df_v2['best_score'].values[:K]-optimal_value
    standard_v3 = Frank_Wolfe_Standard_v3(my_instance)
    np.random.seed(0)
    x_bar_k_v3,df_v3=standard_v3.FW_st_solve(actual_time, analyse=True, K=K, nk=nk,n_pre=n_pre)
    list_v3+=df_v3['best_score'].values[:K]-optimal_value
list_v2=list_v2/nb_test
list_v3=list_v3/nb_test
```

```python
plt.plot([x for x in range(0,200)],list_v2, label="FWstd v2 200 vehicles optimized, 10 tests")
plt.plot([x for x in range(0,200)],list_v3, label="FWstd v3 200 vehicles optimized, 10 tests")
plt.yscale('log')
plt.legend()
plt.title("Evolution of the mean gap between the best score and the optimal value of cplex with 200 vehicles, with 10 tests, 50 FWS, 50 FW Standard")
plt.xlabel("Iteration")
plt.ylabel("Mean gap")
plt.show()
```
